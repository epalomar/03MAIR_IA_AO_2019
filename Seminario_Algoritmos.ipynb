```
## VALOR MÁXIMO Y VALOR MÍNIMO
def calcula_expresiones_2(expresion,cifras,operaciones):
  #print(expresion)
  if len(cifras)==0 or len(operaciones)==0:
    return expresion
  else:
    if expresion=='':
      cifra=choice(cifras)
      cifras.remove(cifra)
      cifra2=choice(cifras)
      cifras.remove(cifra2)
      operacion=choice(operaciones)
      operaciones.remove(operacion)
      expresion=cifra +operacion+cifra2
      return calcula_expresiones_2(expresion,cifras,operaciones)
    else:
      cifra=choice(cifras)
      cifras.remove(cifra)
      operacion=choice(operaciones)
      operaciones.remove(operacion)
      return calcula_expresiones_2(expresion+operacion+cifra,cifras,operaciones)

#Busqueda aleatorio
cifras=['1','2','3','4','5','6','7','8','9']
operaciones=['-','+','*','/']
expresion=''
def busqueda_aleatoria(N):
  mejor_solucion = ""
  mejor_valor    = -1
  peor_solucion = ""
  peor_valor    = -1

  for i in range(N):
      cifras=['1','2','3','4','5','6','7','8','9']
      operaciones=['-','+','*','/']
      expresion=''
      solucion = calcula_expresiones_2(expresion,cifras,operaciones)
      valor    = eval(solucion)
      if valor > mejor_valor:
        mejor_solucion = solucion
        mejor_valor = valor
      if valor < peor_valor:
        peor_solucion = solucion
        peor_valor = valor

  print("El valor más alto es ",mejor_valor,  mejor_solucion)
  print("El valor más bajo es ",peor_valor,  peor_solucion)
  return mejor_valor,peor_valor

maximo,minimo=busqueda_aleatoria(90000)
```

```
## VALOR MÁXIMO MEJORADO
## Se trata de evaluar si el siguiente expresion mejora la anterior,si no la mejora no seguimos avanzando por ese camino
def busqueda_valor_maximo_mejorado(N):
  mejor_solucion = "-9999"
  mejor_valor    = -1
  for i in range(N):
    cifras=['1','2','3','4','5','6','7','8','9']
    operaciones=['-','+','*','/']
    c1=choice(cifras)
    cifras.remove(c1)
    c2=choice(cifras)
    cifras.remove(c2)
    ope=choice(operaciones)
    operaciones.remove(ope)
    expresion=c1+ope+c2
    max=False
    count=len(operaciones)
    iteracion=0
    #print(expresion)
    while ( (len(cifras)!=0) and (len(operaciones)!=0) and max!=True):
        c3=choice(cifras)
        ope=choice(operaciones)
        iteracion=iteracion+1
        expresion_nueva=expresion+ope+c3
        #print(expresion_nueva)
        if (eval(expresion_nueva)>eval(mejor_solucion)):
          mejor_valor=eval(expresion_nueva)
          mejor_solucion=expresion_nueva
          expresion=expresion_nueva
          cifras.remove(c3)
          operaciones.remove(ope)
          count=len(operaciones)
          iteracion=0
        if (iteracion==count):
          max=True
  return mejor_valor,mejor_solucion

busqueda_valor_maximo_mejorado(2000)
```
